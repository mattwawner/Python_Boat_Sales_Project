```python
# Creating an average views column

df['Avg Views'] = df.groupby(['Country'])['Views last week'].transform(np.mean)
```

```python
# Creating a total views column

df['Total Views'] = df.groupby(['Country'])['Views last week'].transform(np.sum)
```

```python
# Creating an average price column

df['Avg Price'] = df.groupby(['Country'])['Euro Price'].transform(np.mean)
```

```python
df.head()
```

```python
columns = ["Country","Avg Views","Total Views","Avg Price"]
```

```python
country_cats=df[columns]
```

### 3. Geographic Analysis

```python
country_cats.head()
```

```python
df1=country_cats.drop_duplicates()
```

```python
df1.head()
```

```python
df1.shape
```

```python
# Setup a folium map at a high-level zoom
map = folium.Map(location = [100, 0], zoom_start = 1.5)

# Choropleth maps bind Pandas Data Frames and json geometries.This allows us to quickly visualize data combinations
folium.Choropleth(
    geo_data = countries, 
    data = df1,
    columns = ['Country', 'Avg Views'],
    key_on = 'feature.properties.name', # this part is very important - check your json file to see where the KEY is located
    fill_color = 'YlOrBr', fill_opacity=0.8, line_opacity=0.1,
    legend_name = "Avg Views").add_to(map)
folium.LayerControl().add_to(map)

map
```
